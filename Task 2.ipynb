## ------API

```python
import pandas as pd
```

```python
import requests

url = "https://maps-data.p.rapidapi.com/searchmaps.php"

querystring = {"query":"school","limit":"200","country":"us","lang":"en","lat":"17.387140","lng":"78.491684","offset":"0","zoom":"13"}

headers = {
	"X-RapidAPI-Key": "169656dc12mshc51db39aae904f7p1f9cadjsnc03266fb5929",
	"X-RapidAPI-Host": "maps-data.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
pd.json_normalize(response.json(), record_path = ['data'])
```
