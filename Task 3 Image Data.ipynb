```python
import cv2
import pandas as pd
import numpy as np
import urllib.request
import os
import requests
```

```python


animals = ['Cat', 'Dog', 'Elephant', 'Parrot', 'Cow']
n= 0
arr= []
label = []

for i in animals:
    url = f"https://free-images-api.p.rapidapi.com/images/{i}"

    headers = {
        "X-RapidAPI-Key": "169656dc12mshc51db39aae904f7p1f9cadjsnc03266fb5929",
        "X-RapidAPI-Host": "free-images-api.p.rapidapi.com"}

    response = requests.get(url, headers=headers)
    data = pd.json_normalize(response.json(),record_path = ['results'] )
    path = os.getcwd()+f'\\{i}'
    os.makedirs(path, exist_ok=True)
    os.chdir(f'G:\Animal Task\\{i}')
    
    
    for j in data['image']:
        urllib.request.urlretrieve(j,f'G:\Animal Task\\{i}\\{i}{n}.jpg')
        n = n+1
    for k in os.listdir():
        path_new = os.getcwd()+f'\\{k}'
        image = cv2.imread(path_new)
        im = cv2.resize(image,(7,6))
        arr.append(im.flatten())
        label.append(i)
dd = pd.DataFrame(arr)
dd['label'] = label
```
